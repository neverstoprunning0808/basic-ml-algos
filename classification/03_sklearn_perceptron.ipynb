```python
# import libraries
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score, accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Perceptron

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# get the data
iris = datasets.load_iris()
X = iris.data
y = iris.target
print(f"Class label: {np.unique(y)}")
print(f"Target name: {iris.target_names}")
```

```python
# split data
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, shuffle=True, test_size=0.2, random_state=8)
```

```python
# check data distro
print(f'Label counts in y: {np.bincount(y)}')
print(f'Label counts in y_train: {np.bincount(y_train)}')
print(f'Label counts in y_test: {np.bincount(y_test)}')
```

```python
# data normalization
scaler = StandardScaler()
X_train_std = scaler.fit_transform(X_train)
X_test_std = scaler.transform(X_test)
```

```python
# get the model, train and predict
perceptron = Perceptron(eta0=0.1, random_state=8)
perceptron.fit(X_train_std, y_train)
y_preds = perceptron.predict(X_test_std)
```

```python
# accuracy score
accuracy = accuracy_score(y_preds, y_test)
print(f"{accuracy = :.4f}")

f1_macro = f1_score(y_test, y_preds, average='macro') # simple average of each binary f1
f1_micro = f1_score(y_test, y_preds, average="micro") # combine TP, TP, FN before computing precision and recall
f1_weighted = f1_score(y_test, y_preds, average='weighted')
print(f"{f1_macro = :.4f}")
print(f"{f1_micro = :.4f}")
```
