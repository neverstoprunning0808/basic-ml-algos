```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# build a vanilla Perceptron class
class Perceptron:
    def __init__(self, eta=0.01, n_iter=50, random_state=1):
        self.eta = eta
        self.n_iter = n_iter
        self.random_state = random_state
    
    def fit(self, X, y):
        random_generation = np.random.RandomState(self.random_state)
        self.w_ = random_generation.normal(size=X.shape[1])
        self.b_ = 0.
        self.errors_ = []

        for i in range(self.n_iter):
            errors = 0.
            for x, target in zip(X, y):
                error = (target - self.predict(x))
                self.w_ += self.eta * error * x 
                self.b_ += self.eta * error
                errors += np.abs(error)
            self.errors_.append(errors)

    def predict(self, X):
        return np.where((np.dot(X, self.w_) + self.b_) >= 0., 1, 0)
```

```python
# get data
source = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
df = pd.read_csv(source, header=None)
df.head()
```

```python
# data exploration
print(df.info())
print(df[4].unique())
```

```python
# get features and labels
features = df.iloc[:,0:4].values
labels = df.iloc[:, 4].values
```

```python
plt.scatter(features[:50, 0], features[:50, 1])
plt.scatter(features[50:100, 0], features[50:100, 1])
plt.scatter(features[100:150, 0], features[100:150, 1])
plt.xlabel('sepal length')
plt.ylabel('sepal width')
plt.legend(['Setosa', 'Versicolor', 'Virginica'])
plt.show()
```

```python
# run the Perceptron algorithm on Setosa and Versicolor
X = features[:100,:]
y = np.where(labels[:100] == "Iris-setosa", 0, 1)
```

```python
# training
perceptron = Perceptron(n_iter=40)
perceptron.fit(X, y)
```

```python
# training errors:
training_errors = perceptron.errors_
n = list(range(perceptron.n_iter))
plt.figure(figsize=(8, 6))
plt.plot(n, training_errors)
plt.show()
```
