#### Basic Pytorch

```python
# create tensor
import torch
import numpy as np

np.set_printoptions(precision=3)

a = [1,2,3]
b = np.array([4,5,6], dtype=np.int32)

tensor_a = torch.tensor(a)
tensor_b = torch.tensor(b)

print(tensor_a, tensor_a.dtype)
print(tensor_b)
```

```python
# create 1s or 0s tensor
t_ones = torch.ones(2, 3)
t_zeros = torch.zeros(2,3)
print(t_ones)
print(t_zeros)
```

```python
# create random tensor
t_rand = torch.rand(2,3)
print(t_rand)
```

```python
# data type and shape of a tensor_a
print(tensor_a.dtype)
tensor_a_new = tensor_a.to(torch.int32)
print(tensor_a_new.dtype)
```

```python
# transpose
t_rand_new = torch.transpose(t_rand, 1, 0)
print(t_rand_new)
```

```python
# reshape
t_zero = torch.zeros(3, 6)
print(t_zero)
t_zero_transpose = t_zero.transpose(1, 0)
print(t_zero_transpose)
t_zero_reshape = t_zero.reshape(-1, 2)
t_zero_reshape_2 = torch.reshape(t_zero, (9, 2))
print(t_zero_reshape)
print(t_zero_reshape_2)
```

```python
# remove dimension: (can only remove dimensions that have size 1)
t = torch.zeros(2,1,2)
print(t, t.shape)
t_squeeze = torch.squeeze(t, 1)
print()
print(t_squeeze, t_squeeze.shape)
```

```python
# add new dimension
t = torch.zeros(2,3)
print(t, t.shape)
t_unsqueeze = torch.unsqueeze(t, 2)
print(t_unsqueeze, t_unsqueeze.shape)
```

```python
# apply mathematical operations:
torch.manual_seed(1)
t1 = 2*torch.rand(5,2) - 1 # uniform distribution from [0,1] -> 2 * [0, 1]  - 1 = [-1, 1]
t2 = torch.normal(mean=0, std=1, size=(5, 2))
print(t1)
print(t2)
```

```python
# element-wise product:
t3 = torch.multiply(t1, t2)
print(t3)
# dot product
t4 = torch.matmul(t1, t2.T)
print(t4)
# mean
t5 = torch.mean(t3, axis=1)
print(t5)
```

```python
t3 = -torch.ones(2, 3)
print(t3)

# compute linalg norm P
norm_t1 = torch.linalg.norm(t3, ord=1, dim=1)
print(norm_t1)

norm_t2 = torch.linalg.norm(t3, ord=2, dim=0)
print(norm_t2)
```

```python
# split
torch.manual_seed(1)
t = torch.rand(6)
print(t)

# split to chunks of 3
t_split = t.split(3)
print(t_split)

# create 3 chunks
t_chunk = t.chunk(3)
print(t_chunk)
```

```python
# concatenate
A = torch.ones(3)
B = torch.zeros(2)
C = torch.concatenate([A, B], axis=0) # same as cat()
print(C)

# stack
D = torch.zeros(3)
S = torch.stack([A, D], axis=0)
print(S)
```
