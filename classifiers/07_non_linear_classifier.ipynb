#### Solving nonlinear problems using a kernel SVM
- SVM is popular among ML algos because they can be easily kernelized to solve nonlinear classification problems.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(1)

X_xor = np.random.randn(200, 2)
y_xor = np.logical_xor(X_xor[:, 0] > 0, X_xor[:, 1] > 0)

y_xor = np.where(y_xor, 1, 0)

plt.scatter(X_xor[y_xor==1, 0],X_xor[y_xor==1, 1], c="royalblue", marker="s", label="Class 1")
plt.scatter(X_xor[y_xor==0, 0],X_xor[y_xor==0, 1], c="tomato", marker="o", label="Class 0")
```

##### The basic idea behind the kernel methods for dealing with such linearly inseparable data is to create nonlinear combination of the original features to project them onto a higher-dimensional space via a mapping function, phi, where data becomes linearly separable.

##### The RBF (radial basis function) kernel is called Gaussian kernel: 
<img src="images/rbf_kernel.png" width=400>

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

svm = SVC(kernel="rbf", random_state=1, gamma=0.1, C=10.0)
svm.fit(X_xor, y_xor)
accuracy = accuracy_score(svm.predict(X_xor), y_xor)
print(accuracy)
```
