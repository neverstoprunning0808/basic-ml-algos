##### K-nearest neighbors: a lazy learning algorithm
- it's lazy because it doesn't learn the algo (no params), but memorizes the trainng data instead.
- miknkowski distance: p and metric \
<img src='images/minkowski_distance.png' width=400>

```python
from sklearn.neighbors import KNeighborsClassifier as KNN
from classification_algos.make_data import generate_iris_data
from sklearn.metrics import accuracy_score
```

```python
knn = KNN(n_neighbors=5, p=2, metric="minkowski")
X_train, X_test, y_train, y_test = generate_iris_data()
knn.fit(X_train, y_train)

preds = knn.predict(X_test)
accuracy = accuracy_score(preds, y_test)
print(accuracy)
```
