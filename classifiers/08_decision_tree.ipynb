```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, f1_score
from classification_algos.make_data import generate_iris_data

tree = DecisionTreeClassifier(criterion="gini", max_depth=4, random_state=1)
X_train, X_test, y_train, y_test = generate_iris_data()

tree.fit(X_train, y_train)
train_preds = tree.predict(X_train)
train_accuracy = accuracy_score(train_preds, y_train)
train_f1_score = f1_score(train_preds, y_train)

test_preds = tree.predict(X_test)
test_accuracy = accuracy_score(test_preds, y_test)
test_f1_score = f1_score(test_preds, y_test)

print(train_accuracy, train_f1_score)
print(test_accuracy, test_f1_score)
```

```python
# Random Forest
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=25, random_state=1, n_jobs=2)
rf.fit(X_train, y_train)


preds = rf.predict(X_test)
accuracy = accuracy_score(preds, y_test)
print(accuracy)
```
